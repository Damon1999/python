## 频率分布直方图

### 简单绘图

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("bili2.csv")
data.head(10)
```

```python
plt.hist(list(data['收盘']),alpha=0.6)#bins和，alpha设置透明度，0为完全透明
#plt.plot()可以绘制折线图，plt.scatter()绘制散点图，plt.bar()绘制直方图，等等
```

### 装饰直方图

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import MultipleLocator #用于刻度标签的间隔
from pylab import mpl #使用中文文字
import pandas as pd
import numpy as np

data = pd.read_csv("bili2.csv")
data.head(10)
```

```python
bins = np.arange(0,50,0.5)
#直方图会进行统计各个区间的数值
plt.hist(list(data['收盘']),bins,alpha=0.6)#bins和，alpha设置透明度，0为完全透明
#颜色可以用'b', 'g', 'r', 'c', 'm', 'y', 'k', 'w'，也可以用'red'，’blue'等，还可以用十六进制

mpl.rcParams['font.sans-serif'] = ['SimHei']#设置中文字体，否则乱码
plt.xlabel('收盘价',fontsize=17)
plt.ylabel('收盘价频率',fontsize=17)
plt.title('频率分布直方图',fontsize=18)

plt.xlim(0.5,50)#设置x轴分布范围
x_major_locator=MultipleLocator(10)#刻度间隔
ax=plt.gca() #获取当前的图表对象
ax.xaxis.set_major_locator(x_major_locator)#x轴主刻度标签
plt.xticks(rotation=60,fontsize=15)

plt.yticks(fontsize=15)

plt.savefig("close_freq.png",dpi=300,bbox_inches = 'tight')
plt.show()
```

# 时间坐标轴设置

```python
import pandas as pd
data = pd.read_csv("bili2.csv")
data = data.sort_values(by=['日期'])
data.head(10)
```

## 时间-错误示范

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import MultipleLocator #用于刻度标签的间隔
from pylab import mpl #使用中文文字

plt.figure(figsize=(10,6))#设置画布大小
plt.plot(data['日期'],data['收盘'],color='#FF76A2',linewidth=2)
#颜色可以用'b', 'g', 'r', 'c', 'm', 'y', 'k', 'w'，也可以用'red'，’blue'等，还可以用十六进制

mpl.rcParams['font.sans-serif'] = ['SimHei']#设置中文字体，否则乱码
plt.xlabel('日期',fontsize=17)
plt.ylabel('收盘价',fontsize=17)
plt.title('B站股价变化',fontsize=18)

#设置x轴标签间隔
x_major_locator=MultipleLocator(100)#刻度间隔
ax=plt.gca() #获取当前的图表对象
ax.xaxis.set_major_locator(x_major_locator)#x轴主刻度标签间隔

plt.xticks(rotation=60,fontsize=15)
plt.yticks(fontsize=15)

plt.savefig("close_freq.png",dpi=300,bbox_inches = 'tight')
plt.show()
```

## 时间-正确示范

```python
#将日期从文本格式转换为时间格式
data['日期'] = pd.to_datetime(data['日期'],format="%Y年%m月%d日")
data = data.sort_values(by=['日期'])
```

```python
#https://matplotlib.org/gallery/text_labels_and_annotations/date_index_formatter.html#sphx-glr-gallery-text-labels-and-annotations-date-index-formatter-py
import matplotlib.pyplot as plt
from matplotlib.pyplot import MultipleLocator #用于刻度标签的间隔
from pylab import mpl #使用中文文字

plt.figure(figsize=(10,6))#设置画布大小
plt.plot(data['日期'],data['收盘'],color='#FF76A2',linewidth=2)

mpl.rcParams['font.sans-serif'] = ['SimHei']#设置中文字体，否则乱码
plt.xlabel('日期',fontsize=17)
plt.ylabel('收盘价',fontsize=17)
plt.title('B站股价变化',fontsize=18)

x_major_locator=MultipleLocator(100)#刻度间隔
ax=plt.gca() #获取当前的图表对象
ax.xaxis.set_major_locator(x_major_locator)#x轴主刻度标签间隔


plt.xticks(rotation=60,fontsize=15)
plt.yticks(fontsize=15)

plt.savefig("close_freq.png",dpi=300,bbox_inches = 'tight')
plt.show()
```

## 时间-正确示范-进阶

```python
#导入matplotlib时间
import matplotlib.dates as mdates
years = mdates.YearLocator() #设置年
months = mdates.MonthLocator()  #设置月
years_fmt = mdates.DateFormatter('%Y') #设置年格式

#https://matplotlib.org/gallery/text_labels_and_annotations/date_index_formatter.html#sphx-glr-gallery-text-labels-and-annotations-date-index-formatter-py
from matplotlib.pyplot import MultipleLocator #用于刻度标签的间隔
from pylab import mpl #使用中文文字

plt.figure(figsize=(10,6))#设置画布大小
plt.plot(data['日期'],data['收盘'],color='#FF76A2',linewidth=2)

mpl.rcParams['font.sans-serif'] = ['SimHei']#设置中文字体，否则乱码
plt.xlabel('日期',fontsize=17)
plt.ylabel('收盘价',fontsize=17)
plt.title('B站股价变化',fontsize=18)

#时间轴坐标刻度间隔设置
ax=plt.gca() #获取当前的图表对象
ax.xaxis.set_major_locator(years)#设置主刻度
ax.xaxis.set_minor_locator(months)#设置次刻度
ax.xaxis.set_major_formatter(years_fmt)#设置标签显示格式

plt.xticks(rotation=60,fontsize=15)
plt.yticks(fontsize=15)

plt.savefig("close_freq.png",dpi=300,bbox_inches = 'tight')
plt.show()
```
